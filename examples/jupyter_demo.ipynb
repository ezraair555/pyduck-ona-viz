# pyduck-ona-viz notebook demo

This notebook walks through four visualizations using a synthetic organization.

```python
import pandas as pd
import pyduck_ona_viz as viz

# Synthetic hierarchy
hierarchy = pd.DataFrame({
    'employee_id': ['CEO', 'VP1', 'VP2', 'M1', 'M2', 'IC1'],
    'supervisor_id': [None, 'CEO', 'CEO', 'VP1', 'VP2', 'M1'],
})
metadata = pd.DataFrame({
    'employee_id': hierarchy['employee_id'],
    'name': [f'Person-{e}' for e in hierarchy['employee_id']],
    'department': ['Exec', 'Eng', 'Sales', 'Eng', 'Sales', 'Eng'],
    'title': ['CEO', 'VP', 'VP', 'Manager', 'Manager', 'IC'],
})
stats = pd.DataFrame({
    'employee_id': ['CEO', 'VP1', 'VP2', 'M1', 'M2'],
    'direct_reports': [2, 2, 1, 1, 0],
    'total_reports': [5, 3, 2, 1, 0],
    'levels_below': [2, 1, 1, 0, 0],
})
print('Data ready')
```

## 1. Span of control

Top managers by direct-report count.

```python
fig = viz.span_of_control(stats, metadata=metadata, top_n=10, color_by_department=True)
fig.savefig('output/notebook_01_span.png', dpi=150, bbox_inches='tight')
print('Saved output/notebook_01_span.png')
```

## 2. Interactive org chart

A collapsible D3 tree coloured by department.

```python
html = viz.org_chart_tree(hierarchy, metadata=metadata, color_by='department')
from pathlib import Path
Path('output/notebook_02_org_chart.html').write_text(html)
print('Saved output/notebook_02_org_chart.html')
```

## 3. Reporting chain walk

Path from any employee up to the top of the org.

```python
fig = viz.reporting_chain_walk(hierarchy, 'IC1', metadata=metadata)
fig.savefig('output/notebook_03_chain.png', dpi=150, bbox_inches='tight')
print('Saved output/notebook_03_chain.png')
```

## 4. Summary dashboard

One-page HTML dashboard with headline KPIs.

```python
html = viz.summary_dashboard(stats)
Path('output/notebook_04_dashboard.html').write_text(html)
print('Saved output/notebook_04_dashboard.html')
```
